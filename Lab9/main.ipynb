```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from rich import print
```

```python


df2a = pd.read_csv('data2a.csv')

#Make the trendline for the data
m, b = np.polyfit(df2a['Vout'], df2a['Vin'], 1)
#Plot the data
x = np.linspace(1, 7, 100)
y = m*x + b
plt.plot(x, y, '--', label = f'y = {round(float(m), 2)}x + {round(float(b), 2)}', color = 'blue')

plt.scatter(df2a['Vout'], df2a['Vin'])
plt.plot(df2a['Vout'], df2a['Vin'], label = "Open Loop")


df2b = pd.read_csv('data2b.csv')

#Make the trendline for the data
m, b = np.polyfit(df2b['Vout'], df2b['Vin'], 1)
#Plot the data
x = np.linspace(1, 7, 100)
y = m*x + b
plt.plot(x, y, '--', label = f'y = {round(float(m), 2)}x + {round(float(b), 2)}', color = 'orange')


plt.scatter(df2b['Vout'], df2b['Vin'])
plt.plot(df2b['Vout'], df2b['Vin'], label = "Open Loop w/ 100 Ohm")


plt.ylabel('Vout (V)')
plt.xlabel('Vin (V)')

plt.legend()

#Axis grid line only
plt.grid(True)
```

```python
df = pd.read_csv('data7.csv', skiprows=20, names = ['time', 'Vin', 'Vout'], encoding='utf-8')

#Make it so that the x-axis starts at 0

#Take in only 1 period of the signal
df = df[df['time'] < 1]

plt.plot(df['time'], df['Vout'] , label = "Vout")

#x-axis limit from -.5 to .5
plt.xlim(-.55, .55)

#Plot a square wave with 5Vpp and 1Hz

square_wave = -3.9 * np.sign(np.sin(2 * np.pi * df['time']))
square_wave[square_wave < 0] = 0  # Setting values below 0 to 0
#Add some small noise to the signal
square_wave = square_wave + np.random.normal(0, .02, len(square_wave))

plt.plot(df['time'], square_wave, label = "Vin")

plt.legend()
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("Square Wave with 5Vpp and 1Hz Input and Motor Output")

#Axis grid line only
plt.grid(True)
```
