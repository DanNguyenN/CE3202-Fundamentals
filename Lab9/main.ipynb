```python
df = pd.read_csv('part7.csv', skiprows=20, names = ['time', 'Vin', 'Vout'], encoding='utf-8')

#Make it so that the x-axis starts at 0

#Take in only 1 period of the signal
df = df[df['time'] < 1]

plt.plot(df['time'], df['Vout'] * 3, label = "Vout")

#x-axis limit from -.5 to .5
plt.xlim(-.55, .55)

#Plot a square wave with 5Vpp and 1Hz

square_wave = -5 * np.sign(np.sin(2 * np.pi * df['time']))
square_wave[square_wave < 0] = 0  # Setting values below 0 to 0

plt.plot(df['time'], square_wave, label = "Vin")

plt.legend()

#Axis grid line only
plt.grid(True)
```
