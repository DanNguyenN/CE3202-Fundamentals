```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from rich import print

#Get data from part1.csv 
df = pd.read_csv('part1.csv')
df.head()

#There are 4 columns in the dataset Time (s), Channel 1 (V), Channel 2 (V), Math 1 (A)

#Plotting the data
plt.figure()
plt.plot( df['Channel 1 (V)'], df['Math 1 (A)'] * 1000, label='IV curve in mA', c = 'k')
plt.xlabel('Voltage (V)')
plt.ylabel('Current (mA)')
plt.title('IV curve')

#Draw the line of x = 2
plt.axvline(x=2, color='k', linestyle='--', label='x = 2')
#Draw the line of y = 4.213
plt.axhline(y=6.87, color='k', linestyle='--', label='y = 4.213')

plt.plot([2], [6.87], 'bo', label='Data points on x = 2')
#Label the point
plt.annotate('(2, 6.77)', xy=(2.2, 7.2), xytext=(2.2, 7.2))


#Calculate the Beta
print(f'Vce: 2 V')
print(f'Ic: 6.87 mA')
print(f'Ib: 40 uA')
print(f'Beta value: {6.87 * 10 ** - 3 / (40 * 10 ** -6)}')
```
