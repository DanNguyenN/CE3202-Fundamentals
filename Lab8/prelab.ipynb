```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Plot the sine wave of 1Khz
t = np.arange(0, 1/1000, 0.000001)
f = 1000
x = 2.5 * np.sin(2*np.pi*f*t) + 2.5
plt.plot(t * 1000,x )
```

```python
#Plot the sine wave of 1Khz
t = np.arange(0, 1/1000, 0.000001)
f = 1000
x = 2.5 * np.sin(2*np.pi*f*t) + 2.5
plt.plot(t * 1000,x )
plt.xlabel('Time (ms)')
plt.xticks([0,1])
plt.yticks([0,10/4,20/4])
plt.title("Approximate Sinusoid with only 3 levels")
plt.grid()
```
