```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1 Bit (INCORRECT, USED 1KHZ / 1000):

```python
digitalLogic = pd.read_csv('.\inlab\\1bit\\1bit_1kHz_DigitalLogic.csv', skiprows=7, encoding='utf-8')
digitalLogic.columns = ['time', 'dio1', 'dio2', 'dio3']
inputOutput = pd.read_csv('.\inlab\\1bit\\1bit_1kHz_InputOutput.csv', skiprows=23, encoding='utf-8')
inputOutput.columns = ['time', 'input', 'output']

#Set the first time point to zero for both data sets
digitalLogic['time'] = digitalLogic['time'] - digitalLogic['time'][0]
inputOutput['time'] = inputOutput['time'] - inputOutput['time'][0]

#Flip the sign and add 1 to dio1 to align with the input and output
digitalLogic['dio1'] = digitalLogic['dio1']* -1 + 1

#Constraint the time to 2 periods
digitalLogic = digitalLogic[digitalLogic['time'] < 4/1000]
inputOutput = inputOutput[inputOutput['time'] < 4/1000]

#Create 2 subplots, small on top and large on bottom
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 5), gridspec_kw={'height_ratios': [.5, 3]} , sharex=True)
ax1.plot(digitalLogic['time'] * 1000, digitalLogic['dio1'], label='dio1')
ax1.set_title('1 Bit - 1Hz Digital Logic')

ax2.plot(inputOutput['time'] * 1000, inputOutput['input'], label='input')
ax2.plot(inputOutput['time'] * 1000, inputOutput['output'], label='output')
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Voltage (V)')

```

# 2 Bits


### 1Hz

```python
inputOutput1 = pd.read_csv('.\inlab\\2bits\\1Hz\\InputOutputPart1.csv', skiprows=23, encoding='utf-8')
inputOutput1.columns = ['time', 'input', 'output']
inputOutput2 = pd.read_csv('.\inlab\\2bits\\1Hz\\InputOutputPart2.csv', skiprows=23, encoding='utf-8')
inputOutput2.columns = ['time', 'input', 'output']
digitalLogic = pd.read_csv('.\inlab\\2bits\\1Hz\\DigitLogic.csv', skiprows=7, encoding='utf-8')
digitalLogic.columns = ['time', 'dio1', 'dio2', 'dio3']

#Limit the time to 3 periods
inputOutput1 = inputOutput1[inputOutput1['time'] < 3]
inputOutput2 = inputOutput2[inputOutput2['time'] < 3]
digitalLogic = digitalLogic[digitalLogic['time'] < 3]

#Set the first time point to zero for both data sets
inputOutput1['time'] = inputOutput1['time'] - inputOutput1['time'][0]
inputOutput2['time'] = inputOutput2['time'] - inputOutput2['time'][0]
digitalLogic['time'] = digitalLogic['time'] - digitalLogic['time'][0]

digitalLogic['dio1'] = digitalLogic['dio1'] * -1 + 1
digitalLogic['dio2'] = digitalLogic['dio2'] * -1 + 1

#Plot the digital logic
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 5), gridspec_kw={'height_ratios': [.5,.5, 3]} , sharex=True)
ax1.plot(digitalLogic['time'], digitalLogic['dio1'], label='dio1')
ax1.set_title('2 Bits - 1Hz Digital Logic')
ax1.set_ylabel('A0          ', rotation=0)

ax2.plot(digitalLogic['time'], digitalLogic['dio2'], label='dio2')
ax2.set_ylabel('A1          ', rotation=0)


#Plot the input and output
ax3.plot(inputOutput1['time'], inputOutput1['input'], label='01')
ax3.plot(inputOutput2['time'], inputOutput2['output'], label='10')
ax3.plot(inputOutput2['time'], inputOutput2['input'], label='11')
ax3.plot(inputOutput1['time'], inputOutput1['output'], label='input')
ax3.set_xlabel('Time (s)')
ax3.set_ylabel('Voltage (V)')
#Set the legend bottom right
ax3.legend(loc='lower right')


```

### 100Hz

```python
inputOutput1 = pd.read_csv('.\inlab\\2bits\\100Hz\InputOutputPart1.csv', skiprows=23, encoding='utf-8')

inputOutput1.columns = ['time', 'input', 'output']
inputOutput2 = pd.read_csv('.\inlab\\2bits\\100Hz\\InputOutputPart2.csv', skiprows=23, encoding='utf-8')
inputOutput2.columns = ['time', 'input', 'output']
digitalLogic = pd.read_csv('.\inlab\\2bits\\100Hz\\DigitLogic.csv', skiprows=7, encoding='utf-8')
digitalLogic.columns = ['time', 'dio1', 'dio2', 'dio3']




#Set the first time point to zero for both data sets
inputOutput1['time'] = inputOutput1['time'] - inputOutput1['time'][0]
inputOutput2['time'] = inputOutput2['time'] - inputOutput2['time'][0]
digitalLogic['time'] = digitalLogic['time'] - digitalLogic['time'][0]

#Shift the input and output to align with the digital logic
inputOutput1['time'] = inputOutput1['time'] + 0.0025
inputOutput2['time'] = inputOutput2['time'] + 0.0025 + 2/1000

#Limit the time to 3 periods
inputOutput1 = inputOutput1[(inputOutput1['time'] < 4/100)]
inputOutput2 = inputOutput2[(inputOutput2['time'] < 4/100)]
digitalLogic = digitalLogic[digitalLogic['time'] < 4/100]

digitalLogic['dio1'] = digitalLogic['dio1'] * -1 + 1
digitalLogic['dio2'] = digitalLogic['dio2'] * -1 + 1

#Make the start at 0.010 seconds
digitalLogic['time'] = digitalLogic['time'] - 0.010
inputOutput1['time'] = inputOutput1['time'] - 0.010
inputOutput2['time'] = inputOutput2['time'] - 0.010

#Plot the digital logic
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 5), gridspec_kw={'height_ratios': [.5,.5, 3]} , sharex=True)
ax1.plot(digitalLogic['time'] * 1000, digitalLogic['dio1'], label='dio1')
ax1.set_title('2 Bits - 100Hz Digital Logic')
ax1.set_ylabel('A0          ', rotation=0)

ax2.plot(digitalLogic['time'] * 1000, digitalLogic['dio2'], label='dio2')
ax2.set_ylabel('A1          ', rotation=0)


#Plot the input and output
ax3.plot(inputOutput1['time'] * 1000, inputOutput1['input'], label='01')
ax3.plot(inputOutput2['time'] * 1000, inputOutput2['output'], label='10')
ax3.plot(inputOutput2['time'] * 1000, inputOutput2['input'], label='11')
ax3.plot(inputOutput1['time'] * 1000, inputOutput1['output'], label='input')
ax3.set_xlabel('Time (ms)')
ax3.set_ylabel('Voltage (V)')
#Set the legend bottom right
ax3.legend(loc='lower right')

#Limit the x axis to 0
ax1.set_xlim(left=0)


```
