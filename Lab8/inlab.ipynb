```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1 Bit (INCORRECT, USED 1KHZ / 1000):

```python
digitalLogic = pd.read_csv('.\inlab\\1bit\\1bit_1kHz_DigitalLogic.csv', skiprows=7 + 500, encoding='utf-8')
digitalLogic.columns = ['time', 'dio1', 'dio2', 'dio3']
inputOutput = pd.read_csv('.\inlab\\1bit\\1bit_1kHz_InputOutput.csv', skiprows=23 + 510, encoding='utf-8')
inputOutput.columns = ['time', 'input', 'output']

#Set the first time point to zero for both data sets
digitalLogic['time'] = digitalLogic['time'] - digitalLogic['time'][0]
inputOutput['time'] = inputOutput['time'] - inputOutput['time'][0]

#Flip the sign and add 1 to dio1 to align with the input and output
inputOutput['input'] = inputOutput['input']* -1 + 4.75
inputOutput['output'] = inputOutput['output']* -1 + 4.75

#Constraint the time to 2 periods
digitalLogic = digitalLogic[digitalLogic['time'] < 3/1000] 
inputOutput = inputOutput[inputOutput['time'] < 3/1000] 

#Compute the root mean square error
rms = np.sqrt(np.mean((inputOutput['input'] - inputOutput['output'])**2))
print('RMS Error: ', rms)


#Create 2 subplots, small on top and large on bottom
fig1, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 5), gridspec_kw={'height_ratios': [.5, 3]} , sharex=True)
ax1.plot(digitalLogic['time'] * 1000, digitalLogic['dio1'], label='dio1', color='black')
ax1.set_title('1 Bit - 1Hz Digital Logic')

ax2.plot(inputOutput['time'] * 1000, inputOutput['input'], label='input', color='black')
ax2.plot(inputOutput['time'] * 1000, inputOutput['output'], label='output', color='red')
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Voltage (V)')

```

### 1Hz

```python
inputOutput1 = pd.read_csv('.\inlab\\2bits\\1Hz\\InputOutputPart1.csv', skiprows=23, encoding='utf-8')
inputOutput1.columns = ['time', 'input', 'output']
inputOutput2 = pd.read_csv('.\inlab\\2bits\\1Hz\\InputOutputPart2.csv', skiprows=23, encoding='utf-8')
inputOutput2.columns = ['time', 'input', 'output']
digitalLogic = pd.read_csv('.\inlab\\2bits\\1Hz\\DigitLogic.csv', skiprows=7, encoding='utf-8')
digitalLogic.columns = ['time', 'dio1', 'dio2', 'dio3']

#Limit the time to 3 periods
inputOutput1 = inputOutput1[inputOutput1['time'] < 1]
inputOutput2 = inputOutput2[inputOutput2['time'] < 1]
digitalLogic = digitalLogic[digitalLogic['time'] < 1]




#Set the first time point to zero for both data sets
inputOutput1['time'] = inputOutput1['time'] - inputOutput1['time'][0]
inputOutput2['time'] = inputOutput2['time'] - inputOutput2['time'][0]
digitalLogic['time'] = digitalLogic['time'] - digitalLogic['time'][0]

#Shift the input and output to align with the digital logic
inputOutput1['time'] = inputOutput1['time']
inputOutput2['time'] = inputOutput2['time'] + .04

digitalLogic['dio1'] = digitalLogic['dio1'] * -1 + 1
digitalLogic['dio2'] = digitalLogic['dio2'] * -1 + 1

#Plot the digital logic
fig1, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 5), gridspec_kw={'height_ratios': [.5,.5, 3]} , sharex=True)
ax1.plot(digitalLogic['time'], digitalLogic['dio1'], label='dio1', color='black')
ax1.set_title('2 Bits - 1Hz Digital Logic')
ax1.set_ylabel('A0          ', rotation=0)

ax2.plot(digitalLogic['time'], digitalLogic['dio2'], label='dio2', color='black')
ax2.set_ylabel('A1          ', rotation=0)



#Plot the input and output
ax3.plot(inputOutput1['time'], inputOutput1['input'], label='01')
ax3.plot(inputOutput2['time'], inputOutput2['output'], label='10')
ax3.plot(inputOutput2['time'], inputOutput2['input'], label='11')
ax3.plot(inputOutput1['time'], inputOutput1['output'], label='input', color = 'red')

# Create a new dataframe with a unified time base
time_base = pd.concat([inputOutput1['time'], inputOutput2['time']]).sort_values().unique()
df = pd.DataFrame({'time': time_base})

#Change the name of the columns to avoid conflicts
inputOutput1.columns = ['time', 'input_01', 'output_01']
inputOutput2.columns = ['time', 'input_02', 'output_02']
# Interpolate signals from inputOutput1 to the unified time base
df = df.merge(inputOutput1, on='time', how='left')

# Interpolating input and output columns
df['input_01'] = df['input_01'].interpolate(method='linear')
df['output_01'] = df['output_01'].interpolate(method='linear')

# Interpolate signals from inputOutput2 to the unified time base
df = df.merge(inputOutput2, on='time', how='left')

# Interpolating input and output columns
df['input_02'] = df['input_02'].interpolate(method='linear')
df['output_02'] = df['output_02'].interpolate(method='linear')

# Calculate the maximum signal at each time point
wave = np.maximum(df['input_02'], df['output_02'] * 0.66)
wave = np.maximum(wave, df['input_01'] * 0.33)

df["output"] = wave

#Calculate the RMS error
rms = np.sqrt(np.mean((df['output_01'] - df["output"])**2))
print('RMS Error: ', rms)
# Plot the wave
ax3.plot(df['time'], wave, label='output', color='black')


ax3.set_xlabel('Time (s)')
ax3.set_ylabel('Voltage (V)')
#Set the legend bottom right
ax3.legend(loc='lower right')


```
