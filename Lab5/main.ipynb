```python
#Plot data in part1.csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('part1.csv', skiprows=11, encoding='latin-1')

plt.scatter(df['Channel 2 (V)'], df['Math 1 (A)'] * 1000, s=.5)
plt.xlabel('VDS (V)')
plt.ylabel('IDS (mA)')

#Plot the line x = 13.59 - 800y
x = np.linspace(0, 14, 100)
y = ((13.59 - x) / 800) * 1000
plt.plot(x, y, color='red')

#Plot the line x = 2.959
x = np.linspace(0, 14, 100)
y = np.linspace(3.267, 3.267, 100)
plt.plot(y, x, color='green')
```
